# 전처리

```python
f = open(r'/content/Curse-detection-data/dataset.txt')
lines = f.readlines()
data_documents = []
data_labels = []
for line in lines:

    line = line.strip()
    label = int(line[-1])
    input = line[:-1].replace("|","")

    data_documents.append(input)
    data_labels.append(label)
f.close()
```

```python
del f
```

```python
from sklearn.model_selection import train_test_split

data_train, data_test,labels_train, labels_test = train_test_split(data_documents, data_labels, test_size=0.1, random_state=42)
```

```python
print(f'train_data:{len(data_train)}')
print(f'train_label:{len(labels_train)}')
print(f'test_data:{len(data_test)}')
print(f'test_label:{len(labels_test)}')
```

```python
data_test[0]
```

```python
import hgtk
documents_train=[]
for i in range(len(data_train)):
    try:
       documents_train.append(list(hgtk.text.decompose(data_train[i], compose_code='/')))
       documents_train[i] = [hangul_mapping[char] for char in documents_train[i]]
    except KeyError:
      print(i)
```

```python
documents_test=[]
for i in range(len(data_test)):
    try:
       documents_test.append(list(hgtk.text.decompose(data_test[i], compose_code='/')))
       documents_test[i] = [hangul_mapping[char] for char in documents_test[i]]
    except KeyError:
      print(i)
```

```python
import matplotlib.pyplot as plt
print('리뷰의 최대 길이 :',max(len(review) for review in documents_train))
print('리뷰의 평균 길이 :',sum(map(len, documents_train))/len(documents_train))
plt.hist([len(review) for review in documents_train], bins=50)
plt.xlabel('length of samples')
plt.ylabel('number of samples')
plt.show()
```

```python
def below_threshold_len(max_len, nested_list):
  count = 0
  for sentence in nested_list:
    if(len(sentence) <= max_len):
        count = count + 1
  print('전체 샘플 중 길이가 %s 이하인 샘플의 비율: %s'%(max_len, (count / len(nested_list))*100))
```

```python
max_len = 500
below_threshold_len(max_len, documents_train)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

documents_train = pad_sequences(documents_train, maxlen=max_len, padding='post')
documents_test = pad_sequences(documents_test, maxlen=max_len, padding='post')
```

```python
import numpy as np
documents_train = np.array(documents_train)
labels_train = np.array(labels_train)
documents_test = np.array(documents_test)
labels_test = np.array(labels_test)
```

# 모델

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Embedding, Dropout, Conv1D, GlobalMaxPooling1D, Dense, Input, Flatten, Concatenate
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.models import load_model

embedding_dim = 16
# dropout_ratio = (0.5, 0.8)
dropout_ratio = 0.3
num_filters = 128
hidden_units = 128
vocab_size = len(hangul_mapping) + 1

kernel_size = 3 # 커널의 크기

num_epochs = 50
```

```python



model = Sequential()
model.add(Embedding(vocab_size, embedding_dim))
model.add(Dropout(dropout_ratio))
model.add(Conv1D(num_filters, kernel_size, padding='valid', activation='relu'))
model.add(GlobalMaxPooling1D())
model.add(Dense(hidden_units, activation='relu'))
model.add(Dropout(dropout_ratio))
model.add(Dense(1, activation='sigmoid'))
#model.add(Dense(3, activation='softmax'))

es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=3)
mc = ModelCheckpoint('best_model.h5', monitor='val_acc', mode='max', verbose=1, save_best_only=True)

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(documents_train, labels_train, epochs=num_epochs, validation_data=(documents_test, labels_test), callbacks=[es, mc])

```

```python
import numpy as np
def preprocess_input(raw_data):
    #print(raw_data)

    decomposed = (list(hgtk.text.decompose(raw_data, compose_code='/')))

    output = [hangul_mapping[char] for char in decomposed]

    padded_seq = np.full(max_len, 0)  # 패딩 값으로 채워진 배열 생성
    padded_seq[:len(output)] = output[:max_len]  # 원본 시퀀스의 값을 복사 (max_len을 초과하는 부분은 잘림)


    out = padded_seq.reshape(1,500)

    return np.array(out)
```
